### Comparing datasets

```python
def get_ticker_set(df):
    tickers = df.select("Symbol").distinct().collect()
    return {str(t["Symbol"]) for t in tickers}

def overlap_heatmap(*tickers):
    tickers = list(tickers)
    N = len(tickers)
    matrix = np.zeros((N, N), dtype=np.intc)
    
    for i in range(N):
        for j in range(N):
            inter = tickers[i].intersection(tickers[j])
            matrix[i][j] = len(inter)
    
    fig, ax = plt.subplots(figsize=(6, 5))
    ticklabels = ["ANN", "NDX", "ETFS", "FUNS"]
    plt.title("Number of overlapping securities per data set", fontsize=18)
    sns.heatmap(matrix, annot=matrix, fmt='d', cmap="RdYlGn", linewidths=0.7, 
                xticklabels=[ticklabels],
                yticklabels=[ticklabels])

tickers_dict = {
    "ANN": get_ticker_set(data_dict["ANN"]),
    "NDX": get_ticker_set(data_dict["NDX"]["stx"]),
    "ETFS": get_ticker_set(data_dict["NDX"]["etfs"]),
    "FUND": get_ticker_set(data_dict["NYSE"]["fundamentals"])
}

overlap_heatmap(*tickers_dict.values())
```

### Pivot stocks so that 1 column = 1 asset
eg. MSFT_Close, MMM_High, JNJ_Volume ...

```python
def pivot(df, on="Date"):
    cols = df.columns[2:]
    df = df.groupBy(on) \
           .pivot("Symbol") \
           .agg(*map(lambda x: f.sum(f.col(x)).alias(x), cols))
    return df

pivot_stocks = pivot(my_stocks)
```

```python
pdf = my_fundamentals.toPandas().drop(["_c0", "PeriodEnding", "ForYear"], axis=1)
trans = pdf.groupby("Symbol").agg(lambda x: list(x))
trans = trans.transpose()

def collapse(row):
    l = pd.Series.tolist(row)
    for i in range(len(l)):
        if not isinstance(l[i], list):
            l[i] = [l[i]]
    flatten = pd.Series([a for b in l for a in b])
    return flatten

trans = trans.apply(collapse, axis=0).fillna(0)
trans.to_csv("fund.csv")
```

### Split data into training and testing periods

```python
def expose_datetime(df):
    return df.withColumn(
        "Date", f.from_unixtime(f.unix_timestamp("Date", "yyyy-MM-dd")).alias("Date"))

def select_period(df, start, end=datetime.now()):
    return df.filter(start < df.Date) \
             .filter(df.Date < end) \
             .sort("Date")

def drop_null_columns(df):
    null_counts = df.select([f.count(f.when(f.col(c).isNull(), c)).alias(c) for c in df.columns]).collect()[0].asDict()
    to_drop = [k for k, v in null_counts.items() if v > 0]
    df = df.drop(*to_drop)
    return df

def print_dim(df):
    print("(Timesteps: {}, Stocks: {})".format(df.count(), len(df.columns)))

# Trading days from 2012 to the present
_t12_20 = select_period(expose_datetime(my_df), 
                        start=datetime(2012, 1, 1))
# Pruning companies that weren't during that period
t12_20 = drop_null_columns(_t12_20)
print_dim(_t12_20)
print_dim(t12_20)

# Train data
t12_16 = select_period(t12_20,
                    start=datetime(2012, 1, 1),
                    end=datetime(2016, 12, 31))

# Test data
t20_20 = select_period(t12_20,
                    start=datetime(2020, 1, 1))
```

## Time series differencing
$$ X_t = \Delta_t - \Delta_0 $$
$$ \Delta_t = S_t - M_t $$
$M_t$ a basket of stocks at time t

$S_t$ the stock price

$X_t$ the differenced signal

```python
def get_cols(df, tags):
    return [f.col(c) for c in df.columns if any([c.endswith("_" + t) for t in tags])]

def add_basket_column(df):
    cols = get_cols(df, ["Close"])
    mean = sum(x for x in cols)/len(cols)
    return df.withColumn('Average_Close', mean)

def difference(df, reference):
    P_0 = df.first()[reference]
    D_0 = ["Date"] + [d - P_0 for d in df.first()[1:]]
    distance_to_mean = lambda r: (r[0], *[r[i] - r[-1] - D_0[i] for i in range(1, len(r))])
    return df.rdd.map(distance_to_mean).toDF(df.columns)


sample_tickers = ["GIS_Close", "MSFT_Close", "MMM_Close", "WMT_Close", "CAT_Close"]
df_sample = t12_16.select(["Date"] + sample_tickers)
df_sample_with_index = add_basket_column(df_sample)

# Plot the portolio, the index separately
pdf_demo = df_sample_with_index.toPandas()
pdf_demo.plot(x="Date", y=sample_tickers, 
    figsize=(15, 5), title="Our Portfolio (semi-log)", logy=True)

pdf_demo.plot(x="Date", y="Average_Close", 
    figsize=(15, 5), title="Stock Market Trend (semi-log)", logy=True)

# Plot the differenced portfolio
differenced = difference(df_sample_with_index, "Average_Close")
differenced.toPandas().plot(
    x="Date", figsize=(15, 5), title="The Differenced Securities (Market Being The Horizontal Line)")
```

### Process the other stocks too

```python
def difference_all(df):
    prices_cols = get_cols(df, ["Open", "Close", "High", "Low"])
    volume_cols = get_cols(df, ["Volume"])
    
    df_volume = df.select(["Date"] + volume_cols)
    df_prices = df.select(["Date"] + prices_cols)    
    
    df_with_basket = add_basket_column(df_prices)
    df_differenced = difference(df_with_basket, "Average_Close")
    
    df_complete = df_volume.join(df_differenced, on="Date", how="outer")
    return df_complete.sort("Date")

diff12_16 = difference_all(t12_16).cache()
diff20_20 = difference_all(t20_20).cache()

diff12_16.toPandas().to_csv("training_pre_minmax.csv")
diff20_20.toPandas().to_csv("testing_pre_minmax.csv")
```

```python
# test_sample = select_period(
#     add_basket_column(t12_16.select(["Date"] + sample_tickers)),
#     start=datetime(2016, 11, 1)
# ).sort("Date")

def vector_assembler(df):
    assembler = VectorAssembler(
        inputCols=df.columns[1:],
        outputCol="features"
    )
    output = assembler.transform(df).select("Date", "features")
    return output

def disassembler(df, schema):
    extract = lambda row: (row.Date,) + tuple(row.scaledFeatures.toArray().tolist())
    df = df.rdd.map(extract).toDF(schema)
    return df

def min_maxer(df):
    scaler = MinMaxScaler(
        inputCol="features", 
        outputCol="scaledFeatures"
    )
    scalerModel = scaler.fit(df)
    scaledData = scalerModel.transform(df)
    return scaledData.select("Date", "scaledFeatures")

def un_min_mixer(df, model):
    raise NotImplementedError

def compose(df, *transformations):
    for f in tqdm(list(transformations)):
        df = f(df)
    return df

vec_training_set = compose(diff12_16, vector_assembler, min_maxer)
vec_testing_set = compose(diff20_20, vector_assembler, min_maxer)

training_set = disassembler(vec_training_set, diff12_16.columns)
testing_set = disassembler(vec_testing_set, diff20_20.columns)
```
