The hope is that this very plain model for how prices move captures phenomena such as normality, momentum, etc.

The first round of the analysis is spent adjusting this general model to the 2012-2016 time period. In particular, to train it on price differences with the market in order to build an investment strategy (call individual stock / short market and vice versa).

```python
import sys
import numpy as np
import tensorflow as tf
import tensorflow.keras as keras
from jupyterthemes import jtplot
from pyspark.sql import SparkSession
from pyspark.ml.linalg import Vectors

from tqdm import tqdm
import pandas as pd
import matplotlib.pyplot as plt
import optuna
import plotly.express as px

jtplot.style()
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("covid19") \
    .getOrCreate()

print(sys.version, "\n", tf.__version__)
```

## Load 2012-2016 Data

```python
def split_data(df, split_time):
    series_pdf = df.toPandas()
    
    series, time = series_pdf.iloc[:, 1:], series_pdf.iloc[:, :1]
    t_train, t_valid = time[:split_time], time[split_time:]
    x_train, x_valid = series[:split_time], series[split_time:]
    return (x_train, t_train), (x_valid, t_valid)

def windowed_dataset(dataset, window_size, batch_size, shuffle_buffer):
    dataset = tf.data.Dataset.from_tensor_slices(dataset)
    dataset = dataset.window(window_size + 1, shift=1, drop_remainder=True)
    dataset = dataset.flat_map(lambda window: window.batch(window_size + 1))
    dataset = dataset.shuffle(shuffle_buffer).map(lambda window: (window[:-1], window[-1:]))
    dataset = dataset.batch(batch_size).prefetch(1)
    return dataset
```

## Setup

```python
## Hyperparameter cell
batch_size = 32
lr = 0.0022571
momentum = 0.9
epochs = 150
lstm1, lstm2 = 128, 128
```

```python
window_size = 60
split_time = 950
num_cols = 1521
shuffle_buffer_size = 1000

df = spark.read.csv("training_set.csv", header=True, inferSchema=True).drop("_c0")
(x_train, t_train), (x_valid, t_valid) = split_data(df, split_time)

print(x_train.shape)
dataset = windowed_dataset(x_train, window_size, batch_size, shuffle_buffer_size)
```

## Model building

```python
## Utility callbacks
path_checkpoint = "model_checkpoint.h5"
es_callback = keras.callbacks.EarlyStopping(monitor="val_loss", 
                                            min_delta=0, patience=5)

modelckpt_callback = keras.callbacks.ModelCheckpoint(
    monitor="val_loss",
    filepath=path_checkpoint,
    verbose=1,
    save_weights_only=True,
    save_best_only=True
)

lr_callback = keras.callbacks.LearningRateScheduler(
    lambda epoch: 1e-8 * 10 ** (epoch/20))
```

```python
def create_model(lstm1, lstm2):
    ## Vanilla
    vanilla = tf.keras.models.Sequential([
        keras.Input(shape=(window_size, num_cols)),
        keras.layers.LSTM(lstm1, name="lstm0"),
        keras.layers.Dense(num_cols, name="l0")
    ])

    ## Bi-LSTM
    bilstm = tf.keras.models.Sequential([
        keras.Input(shape=(window_size, num_cols)),
        keras.layers.Conv1D(filters=32, kernel_size=5,
                            strides=1, padding="causal",
                            activation="relu"),
        keras.layers.LSTM(lstm1, return_sequences=True),
        keras.layers.LSTM(lstm2),
        keras.layers.Dense(num_cols),
        keras.layers.Lambda(lambda x: x * 2.0)
    ])

    return bilstm

def compile_model(model, lr):
    optimizer = keras.optimizers.SGD(lr=lr, momentum=momentum)
    model.compile(
        loss=keras.losses.Huber(),
        optimizer=optimizer,
        metrics=["mae", "accuracy"]
    )
    return my_model
    
def fit_model(model, epochs, V):
    hist = model.fit(dataset, epochs=epochs, verbose=V)
    return hist
```

## Training

```python
my_model = create_model(lstm1, lstm2)
my_model = compile_model(my_model, lr)
history = fit_model(my_model, epochs, V=1)
```

## Validation, Tuning

```python
# Optuna Hyperparameter Tuning
def objective(trial):
    lr=trial.suggest_loguiform("lr", 1e-6, 1e-1)
    lstm1=trial.suggest_categorical("lstm1", [64, 128, 256, 512])
    lstm2=trial.suggest_categorical("lstm2", [128, 256, 512])
    
    my_model = create_model(lstm1, lstm2)
    my_model = compile_model(my_model)
    hist = fit_model(my_model, V=0)
    
    return hist.history["accuracy"][0]

def visualize_optuna(study):
    cols = ["value", "duration", "params_lr", "params_lstm1", "params_lstm2"]
    opt_df = study.trials_dataframe()[cols].sort_values(by=["value"], ascending=False)
    opt_df = opt_df.rename(columns={
        "value": "accuracy",  "params_lr": "LR", 
        "params_lstm1":  "LSTM_1", 
        "params_lstm2": "LSTM_2"})

    fig = px.scatter_3d(opt_df, x="LSTM_1", y="LSTM_2", z="accuracy", color="LR", template="ggplot2" )
    fig.show()


# study = optuna.create_study(direction="maximize")
# study.optimize(objective, n_trials=50)
# visualize_optuna(study)
```

## Forecasting feedback

```python
def forecast(model):
    forecast = []
    for time in tqdm(range(len(x_train) - window_size)):
        sl = x_train[time:time + window_size].to_numpy().reshape(-1, window_size, num_cols)
        forecast.append(my_model.predict(sl))
        
    fc = np.stack(forecast).reshape(-1, 1521)
    pdf_fc = pd.DataFrame(fc, columns=x_train.columns)
    
    sample = ["AAL_Close", "MSFT_Close", "MMM_Close"]
    fig = plt.figure()
    for s in sample:
        ax = x_train.plot(y=s, figsize=(15, 5), label="$\^{y}\_act$",
                          title="1-Step Prediction: " + s)
        pdf_fc.plot(y=s, ax=ax, label="$y\_pred$")
    
    err = keras.metrics.mean_absolute_error(forecast, x_valid[20:]).numpy().mean()
    print("MAE for 930 predictions = %.3f" % err)
    return forecast
 

forecast = forecast(my_model)
```
